```python
import pandas as pd
import requests
import pymysql
import json
from bs4 import BeautifulSoup
from sqlalchemy import create_engine
from selenium import webdriver
```

```python
host_name = "127.0.0.1"
username = "root"
password = ""
database_name = "steam_test"
```

```python
db = pymysql.connect(
    host = host_name,
    port=3306,
    user=username,
    passwd=password,
    db=database_name,
    charset='utf8'
)
```

```python
# 크롤링
id_name = {}
def crolling(count , target):
    while count <= target:
        global error_count
        error_count = count
        res = requests.get('https://steamdb.info/app/{}'.format(count))
        soup = BeautifulSoup(res.content, 'html.parser')
        name_tag = soup.find('h1', class_='css-truncate')
        name = name_tag.text
        
        if name != 'None':
            id_name[count] = name
            count += 1
        else:
            count += 1
    print(id_name)

    try:
        cursor = db.cursor()
        for k, v in id_name.items():
            appid_name_sql = ("""
                INSERT INTO steam_test.steamtable(
                    appid,
                    name
                ) VALUES (
                    {},
                    "{}"
                ) 
                ON DUPLICATE KEY UPDATE appid={}, name="{}";
            """.format(k,v,k,v))
            cursor.execute(appid_name_sql)
            db.commit()
    finally:
        pass
                
    return count, target

# 예외처리
def catch_ex(count, target):
    try:
        crolling(count, target)
    except:
        print("error_count : {}" .format(error_count))
        print("예외 발생 count: {}, target: {}".format(count, target))
        count = error_count + 1   
        catch_ex(count, target)

        
# catch_ex(570, 1000)
# catch_ex(1001,1500)

start_num = 1700
end_num = 1800
# 너무 많은 숫자를넣으면 커널이 끊김으로 짤라서 넣어본다.
for i in range(3):
    print("시작 숫자 :{} 마지막 숫자 :{}".format(start_num,end_num))
#     print(type(start_num), type(end_num))
    catch_ex(start_num, end_num)
    start_num = end_num
    end_num = end_num + 100
```
