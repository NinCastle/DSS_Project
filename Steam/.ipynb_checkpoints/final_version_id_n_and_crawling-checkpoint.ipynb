```python
# ID, name 크롤링
id_name = {}
def crolling(count , target):
    while count <= target:
        global error_count
        error_count = count
        res = requests.get('https://steamdb.info/app/{}'.format(count))
        soup = BeautifulSoup(res.content, 'html.parser')
        name_tag = soup.find('h1', class_='css-truncate')
        name = name_tag.text
        
        if name != 'None':
            id_name[count] = name
            count += 1
        else:
            count += 1
    print(id_name)

    try:
        cursor = db.cursor()
        for k, v in id_name.items():
            appid_name_sql = ("""
                INSERT INTO steam_test.steamtable(
                    appid,
                    name
                ) VALUES (
                    {},
                    "{}"
                ) 
                ON DUPLICATE KEY UPDATE appid={}, name="{}";
            """.format(k,v,k,v))
            cursor.execute(appid_name_sql)
            db.commit()
    finally:
        pass
                
    return count, target

# 예외처리
def catch_ex(count, target):
    try:
        crolling(count, target)
    except:
        print("error_count : {}" .format(error_count))
        print("예외 발생 count: {}, target: {}".format(count, target))
        count = error_count + 1   
        catch_ex(count, target)
        
# 너무 많은 숫자를넣으면 커널이 끊김으로 짤라서 넣어본다.
# 마지막 아이디 : 999999
start_num = 133900
end_num = 134000

# start_num = 10
# end_num = 20

for i in range(50):
    print()
    print()
    print("시작 숫자 :{} 마지막 숫자 :{} for문 횟수 :{}".format(start_num,end_num, i+1))
    catch_ex(start_num, end_num)
    start_num = end_num
    end_num = end_num + 100
```
